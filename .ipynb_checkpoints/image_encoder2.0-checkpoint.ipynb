```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import random
```

```python
def image_encoder(im_array, border=5, random_seed=88, steps=1):
    im_array = im_array
    border = border
    seed = random_seed
    steps = steps
    random.seed(seed)
    
    for w in range(1, im_array.shape[0], steps):        
        for h in range(1, im_array.shape[1], steps):
            for c in range(im_array.shape[2]):
                intensity_b = random.randint(230, 235)
                intensity_w = random.randint(10, 15)
                if im_array[w,h,c] > 200:
                    im_array[w,h,c] = intensity_b
                else:
                    im_array[w,h,c] = intensity_w
                    
    return im_array

def img_plotter(image, display=True, fig_size=10, output_folder='encoded_img'):
    display = display
    image = image
    fig_size = fig_size
    output_folder = output_folder
    if fig_size == None:
        imgplot = plt.figure()
    else:
        imgplot = plt.figure(figsize=(fig_size,fig_size))
    imgplot = plt.axis('off')
    imgplot = plt.imshow(image)
    imgplot = plt.savefig(output_folder)
```

```python
SRC_IMG_PATH = 'qr-files/'
DST_IMG_PATH = 'encoded-files/'

for image in os.listdir(SRC_IMG_PATH):  
    I = np.asarray(Image.open(os.path.join(SRC_IMG_PATH, image)))
    I2 = np.copy(I)
    if len(I2.shape) <  3:
        print('WARNING: Imagem {0}  2D, skipping!'.format(image))
    else: 
        I2 = image_encoder(I2, random_seed=87, steps=1)
        im2 = Image.fromarray(np.uint8(I2))
        img_plotter(im2, fig_size=None, output_folder=DST_IMG_PATH+'10'+image)
```
